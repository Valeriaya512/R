## зҹ©йҷЈйҒӢз®—


*   A * B : зҹ©йҷЈAиҲҮзҹ©йҷЈBзҡ„е…ғзҙ й–“е…©е…©зӣёд№ҳ
*   A %*% B : зҹ©йҷЈд№ҳжі•A вҖ§ B
*   t(A) : Aзҡ„иҪүзҪ®зҹ©йҷЈ
*   solve(A) : зҹ©йҷЈAзҡ„еҸҚзҹ©йҷЈ
*   solve(A,b) : жұӮз·ҡжҖ§зі»зөұAx = bзҡ„и§Ј
*   y <- eigen(A) : yпј„val зӮәAзҡ„зү№еҫөеҖјпјҢyпј„vec зӮәAзҡ„зү№еҫөеҗ‘йҮҸ
*   y <- svd(A) : yпј„u , yпј„d , yпј„v зӮәеҘҮз•°еҖјеҲҶи§ЈA = UD(Vзҡ„иҪүзҪ®)зҡ„зөҗжһң





```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
B <- matrix( c(1, 1, 3, 3), nrow = 2, ncol = 2)
print(A)
print(B)
```

```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
B <- matrix( c(1, 1, 3, 3), nrow = 2, ncol = 2)
print(A * B)
```

```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
B <- matrix( c(1, 1, 3, 3), nrow = 2, ncol = 2)
print(A %*% B)
```

```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
print(t(A))
```

```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
print(solve(A))
```

```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
b <- c(2, 1)
print(solve(A,b))
```

```r
A <- matrix(c(2, 1, 1, 2), nrow = 2, byrow = TRUE)
eig <- eigen(A)

print(eig$values)
print(eig$vectors)

#check
print(eig$vectors %*% diag(eig$values) %*% solve(eig$vectors))
```

```r
A <- matrix(c(1, 2, 3, 4, 5, 6), nrow = 2, byrow = TRUE)

s <- svd(A)

print(s$u)
print(s$d)
print(s$v)

#check
print(s$u %*% diag(s$d) %*% t(s$v))
```
