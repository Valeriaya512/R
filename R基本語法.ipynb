```r
a <- 2:9
a[3] <- 3 * a[3] #將第3個數值*3並取代
print(a)
```

## 矩陣運算


*   A * B : 矩陣A與矩陣B的元素間兩兩相乘
*   A %*% B : 矩陣乘法A ‧ B
*   t(A) : A的轉置矩陣
*   solve(A) : 矩陣A的反矩陣
*   solve(A,b) : 求線性系統Ax = b的解
*   y <- eigen(A) : y＄val 為A的特徵值，y＄vec 為A的特徵向量
*   y <- svd(A) : y＄u , y＄d , y＄v 為奇異值分解A = UD(V的轉置)的結果





```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
B <- matrix( c(1, 1, 3, 3), nrow = 2, ncol = 2)
print(A)
print(B)
```

```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
B <- matrix( c(1, 1, 3, 3), nrow = 2, ncol = 2)
print(A * B)
```

```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
B <- matrix( c(1, 1, 3, 3), nrow = 2, ncol = 2)
print(A %*% B)
```

```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
print(t(A))
```

```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
print(solve(A))
```

```r
A <- matrix( c(1, 2, 3, 4), nrow = 2, ncol = 2)
b <- c(2, 1)
print(solve(A,b))
```

```r
A <- matrix(c(2, 1, 1, 2), nrow = 2, byrow = TRUE)
eig <- eigen(A)

print(eig$values)
print(eig$vectors)

#check
print(eig$vectors %*% diag(eig$values) %*% solve(eig$vectors))
```

```r
A <- matrix(c(1, 2, 3, 4, 5, 6), nrow = 2, byrow = TRUE)

s <- svd(A)

print(s$u)
print(s$d)
print(s$v)

#check
print(s$u %*% diag(s$d) %*% t(s$v))
```
