## 自訂函數
*目的：避免複製大量重複的程式碼*

一個函數組成包含：

*   Input變數x
*   Output變數y
*   運算程式碼

f <- function(x){運算程式碼； return (y)}

```r
# 未自訂函數

df <- data.frame(
    a = rnorm(10), #rnorm：產生10個來自常態分配的隨機數，預設平均數是0、標準差是1
    b = rnorm(10),
    c = rnorm(10),
    d = rnorm(10)
)

df$a <- (df$a - min(df$a, na.rm = TRUE)) /  # na.rm = TRUE 表示忽略缺失值
  (max(df$a, na.rm = TRUE) - min(df$a, na.rm = TRUE))
df$b <- (df$b - min(df$b, na.rm = TRUE)) /
  (max(df$b, na.rm = TRUE) - min(df$b, na.rm = TRUE))
df$c <- (df$c - min(df$c, na.rm = TRUE)) /
  (max(df$c, na.rm = TRUE) - min(df$c, na.rm = TRUE))
df$d <- (df$d - min(df$d, na.rm = TRUE)) /
  (max(df$d, na.rm = TRUE) - min(df$d, na.rm = TRUE))

df
```

####自訂函數注意事項：

1.   **函數要簡短清楚**：
    一個函數最好只負責一件事，太長會很難讀，也不好除錯
2.   **參數名稱要有意義**：
    像 x、y 可以用在簡單範例，但正式寫程式時，盡量用看得懂的名稱
3.   **避免依賴外部變數**：
    函數最好只靠輸入參數運作，不要偷偷用工作區裡的變數
4.   **要處理錯誤輸入**：
    例如空向量、非數值、缺失值，最好先檢查再運算
5.   **回傳結果要固定**：
    同一個函數在不同情況下，回傳型態最好一致，方便後續使用
6.   **不要重複寫程式**：
    把常用的流程包成函數，可以減少重複，也比較容易修改


```r
df <- data.frame(
    a = rnorm(10),
    b = rnorm(10),
    c = rnorm(10),
    d = rnorm(10)
)

RescaleByRange <- function(x){
  rng <- range(x, na.rm = TRUE) #先找出x的最小值和最大值
  return((x - rng[1]) / (rng[2] - rng[1])) #把每個值轉成0到1之間，然後回傳
}

#比較兩者之間結果
(df$a - min(df$a, na.rm = TRUE)) /
  (max(df$a, na.rm = TRUE) - min(df$a, na.rm = TRUE)) #將值轉換成0～1

RescaleByRange(df$a)
```

AND

*   &
*   &&

OR

*   |
*   ||

備註：

*   & 和 |：適合處理「一整串」布林值
*   && 和 ||：適合寫在 if 裡，只檢查第一個值




| 運算子 | 意思  | 比較方式     | 常見用途    | 範例                              |
| --- | --- | -------- | ------- | ------------------------------- |
| &   | AND | 逐一比較整個向量 | 向量條件判斷  | c(TRUE, FALSE) & c(TRUE, TRUE)  |
| &&  | AND | 只看第一個元素  | if 條件判斷 | TRUE && FALSE                   |
| \|  | OR  | 逐一比較整個向量 | 向量條件判斷  | c(TRUE, FALSE) \| c(FALSE, TRUE) |
| \|\| | OR  | 只看第一個元素  | if 條件判斷 | FALSE \|\| TRUE                 |

```r
age <- 18
has_id <- TRUE

age >= 20 & has_id
```

```r
is_student <- FALSE
is_member <- TRUE

is_student | is_member
```

```r
age <- 20
has_id <- TRUE

age >= 18 && has_id
```

```r
is_student <- FALSE
is_member <- FALSE

is_student || is_member
```

```r
set.seed(0)
x <- runif(8, -1,1)

x
```

```r
0 >= x & x <= 0.5
```

```r
#0 >= x && x <= 0.5 #因x是向量，&&只能處理長度為1的邏輯值

#解決辦法一
print(
  if (all(x >= 0) && all (x <= 0.5)){
    "全部都在範圍內"
  }
)

#解決辦法二
print(
  if (any (x >= 0) && any (x <= 0.5)){
    "至少有一個值在範圍內"
  }
)
```

```r
#x是單一數值
x <- -0.3

x >= 0 && x <= 0.5
x >= 0 || x <= 0.5
```

```r
set.seed(0)
x <- runif(8, -1,1)

x

-0.5 >= x | x <= 0.5
```

```r
set.seed(0)
x <- runif(8, -1,1)

x
print(
  if (any (x < -0.5) || any (x > 0.5)){
    "有值超出範圍"
  }
)
```

### set.seed()：固定隨機數
常見搭配：


*   runif：均勻分配隨機數
*   rnorm：常態分配隨機數



```r
set.seed(0)

u <- runif(5)
n <- rnorm(5)

u
n
```

```r
set.seed(0)  #0：決定起始點。起始點可以自行決定
x <- runif(8, -1, 1)  #產生8個介於-1到1之間的隨機數
y <- rnorm(5, 50, 10)  #產生5個平均數50、標準差10的隨機變數

x
y
```

```r
#整合範例
set.seed(0)

df <- data.frame(
  uniform = runif(5),
  normal = rnorm(5)
)

df
```
